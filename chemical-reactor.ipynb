#### Task 4: Simulation of a Chemical Reactor
##### Problem Statement - There are two substances A and B. They produce third substance C. The rate of formation of C is proportional to presence of A and B. Write a program to simulate how much of C has been produced as a function of time, âˆ†t = 0.1. Assume that A = 100g, B = 50g and C = 0g are quantities of A, B and C at time t = 0. Rate constants are also given as follows k1 = 0.008 and k2 = 0.002.

### Chemical Reactor Simulation

```python
import matplotlib.pyplot as plt

A, B, C, k1, k2 = 100, 50, 0, 0.008, 0.002
time, end_time, dt = 0, 1, 0.1

substance_A = []
substance_B = []
substance_C = []
time_points = []

while True:
    substance_A.append(A)
    substance_B.append(B)
    substance_C.append(C)
    time_points.append(time)
    time += dt
    if time > end_time:
        break
    A = A + (k2 * C - k1 * A * B) * dt
    B = B + (k2 * C - k1 * A * B) * dt
    C = C + (2 * k1 * A * B - 2 * k2 * C) * dt
    
plt.figure(figsize = (13, 7))
plt.title("Chemical Reactor Simulation")
plt.xlabel("Time")
plt.ylabel("Quantity")
plt.plot(time_points, substance_A, label = "Substance A")
plt.plot(time_points, substance_B, label = "Substance B")
plt.plot(time_points, substance_C, label = "Substance C")
plt.legend()
plt.grid()
plt.show()
```
