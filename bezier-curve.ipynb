#### Task 4: Simulation of a Cubic Bezier Curve

##### Problem Statement - There are four control points in 2D space. You have to write a program to construct the spline using Cubic Bezier Curve construction approach. Also need to display the output for each new point.

### Function for Factorial

```python
def factorial(n):
    if n:
        return n * factorial(n - 1)
    else:
        return 1
```

### Function for nCr

```python
def nCr(n, r):
    return factorial(n) / float(factorial(r) * factorial(n - r))
```

### Bezier Blending Function

```python
def BEZ(k, n, u):
    return nCr(n, k) * (u ** k) * ((1 - u) ** (n - k));
```

### Function for Bezier Curve

```python
def bezierCurve(x_control_points, y_control_points):
    n = len(x_control_points) - 1
    eps = 0.001
    x_curve_points = []
    y_curve_points = []
    u = 0

    while True:
        x, y = 0, 0
        for k in range(0, len(x_control_points)):
            bez = BEZ(k, n, u)
            x += x_control_points[k] * bez
            y += y_control_points[k] * bez
        x_curve_points.append(x)
        y_curve_points.append(y)
        u += eps
        if u > 1:
            break

    import matplotlib.pyplot as plt
    plt.plot(x_control_points, y_control_points)
    plt.plot(x_curve_points, y_curve_points)
```

### Generating a Cubic Bezier Curve

```python
x_control_points = [27, 128, 327, 437]
y_control_points = [243, 94, 231, 101]

bezierCurve(x_control_points, y_control_points)
```
